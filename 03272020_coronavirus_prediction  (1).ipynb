# India

```r
 df_IN <- df1 %>% filter(Country %in% c('India'))
```

```r
 head(df_IN)
```

```r
j <- ggplot(df_IN, aes(ObservationDate, Confirmed, group = 1))
j + geom_line() + 
  labs(x = "Date", y = "Confirmed cases", 
              title = "Coronavirus Condirmed cases in India") + theme(axis.text.x = element_text(angle = 90, size = 5))
```
